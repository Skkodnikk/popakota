```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = {
    '1': np.random.randint(1, 11, size=5),
    '2': np.random.randint(0, 101, size=5),
    '3': np.random.randint(5000, 35001, size=5),
    '4': np.random.uniform(0.1, 0.5, size=5)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.hist(bins=10, figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
df_stand = df.copy()
df_norm = df.copy()
```

```python
from sklearn.preprocessing import StandardScaler
scalers = StandardScaler(). fit(df_stand)
stand = scalers. transform(df_stand)
stand
```

```python
stand_df = pd.DataFrame(stand)
stand_df
```

```python
stand_df.hist(bins=10, figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import Normalizer
scalern = Normalizer(). fit(df_norm)
norm = scalern. transform(df_norm)
norm
```

```python
norm_df = pd.DataFrame(norm)
norm_df
```

```python
norm_df.hist(bins=10, figsize=(10, 8))
plt.tight_layout()
plt.show()
```
