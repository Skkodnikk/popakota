```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/youth_smoking_drug_data_10000_rows_expanded.csv')
df.head()
```

```python
# Гистограмма (Histplot)
sns.histplot(x='Age_Group', data=df, color='violet')
plt.title('Распределение возраста пользователей')
plt.show()
```

```python
sns.histplot(x='Age_Group', data=df, hue='Gender', palette='pastel')
plt.title('Распределение возраста пользователей по полу')
plt.show()
```

```python
# Boxplot
sns.boxplot(x='Year', y='Drug_Experimentation', data=df, palette='pastel')
plt.title('Эксперименты с наркотиками по годам')
plt.show()
```

```python
sns.violinplot(x='Year', y='Drug_Experimentation', data=df, palette='pastel')
plt.title('Эксперименты с наркотиками по годам (Violinplot)')
plt.show()
```

```python
# Диаграмма рассеяния (Scatter Plot)
sns.scatterplot(x='Drug_Experimentation', y='Age_Group', data=df, hue='Access_to_Counseling', palette='pastel')
plt.title('Зависимость экспериментов с наркотиками от возрастной группы')
plt.show()
```

```python
sns.scatterplot(x='Drug_Experimentation', y='Peer_Influence', data=df, size='Socioeconomic_Status', hue='Access_to_Counseling', palette='pastel')
plt.title('Зависимость экспериментов с наркотиками от влияния (размер точки - социальное положение)')
plt.show()
```

```python
# Lineplot - отображает зависимость между двумя переменными, одна из которых непрерывная, а другая - категориальная
sns.lineplot(x='Mental_Health', y='Age_Group', data=df, color='violet')
plt.title('Зависимость ментального здоровья от возрастной группы')
plt.show()
```

```python
sns.lineplot(x='Age_Group', y='Smoking_Prevalence', data=df, hue='Socioeconomic_Status', palette='pastel')
plt.title('Зависимость возрастной группы от распространенности курения (по социальному статусу)')
plt.show()
```
