```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

X = np.random.uniform(-10, 10, 100)
nor = np.random.normal(0, 2, 100)
Y = 4 * X + 8 + nor

plt.scatter(X, Y, color='blue', label='Данные с шумом')
plt.plot(X, 4 * X + 8, color='red', label='Линейная зависимость', linewidth=2)
plt.title('данные для y=4x+8 с шумом')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X_train = X[0::2].reshape(-1,1)
Y_train = Y[0::2]
X_test = X[1::2].reshape(-1,1)
Y_test = Y[1::2]
plt.scatter(X_train, Y_train)
plt.show()
```

```python
def fit(X, Y):
  n, k = X.shape
  X = np.hstack((X, np.ones((n, 1))))
  w = np.linalg.inv(X.T @ X) @ X.T @ Y
  return w

def predict(X, w):
  n, k = X.shape
  X = np.hstack((X, np.ones((n, 1))))
  Y_pred = X @ w
  return Y_pred
```

```python
W1 = fit(X_train, Y_train)
Y_hat1 = predict(X_train, W1)
W2 = fit(X_test, Y_test)
Y_hat2 = predict(X_test, W2)
plt.hist((Y_hat1 - Y_hat2)**2, bins = 20)
plt.show()
```
