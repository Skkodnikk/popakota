В данном датасете необходимо масштабировать столбцы "Amount" и "Time"

```python
df.hist(figsize=(15, 15))
```

```python
plt.figure(figsize=(12, 9))
plt.subplot(2, 2, 1)
plt.hist(df['Time'], bins=50)
plt.title('Распределение Time')
plt.xlabel('Время')
plt.ylabel('Количество транзакций')

plt.subplot(2, 2, 2)
sns.distplot(df['Time'], kde=False)
plt.title('Распределение Time')
plt.xlabel('Время')
plt.ylabel('Количество транзакций')

plt.subplot(2, 2, 3)
plt.hist(df['Amount'], bins=50)
plt.title('Распределение Amount')
plt.xlabel('Сумма')
plt.ylabel('Количество транзакций')

plt.subplot(2, 2, 4)
sns.distplot(df['Amount'], kde=True)
plt.title('Распределение Amount')
plt.xlabel('Сумма')
plt.ylabel('Плотность')
```

Построили графики распределения для столбцов: Time и Amount.

```python
from sklearn.preprocessing import RobustScaler
rob_scaler = RobustScaler()
df['Time1'] = rob_scaler.fit_transform(df['Time'].values.reshape(-1,1))
df['Amount1'] = rob_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
```

В данном датасете необходимо масштабировать столбцы "Amount" и "Time".

```python
df.head()
```

```python
df1= df.drop(['Time', 'Amount'], axis=1)
df1.head()
```

```python
df1.isnull().sum()
```

Делаем вывод, что пустых значений нет.

```python
sub_sample_corr = df1.corr()
plt.figure(figsize=(30, 15))
sns.heatmap(sub_sample_corr, annot=True, cmap='coolwarm')
plt.show()
```

Выводы:
Отрицательные корреляции:
Признаки V2 и V5 показывают отрицательную корреляцию. Это означает, что при снижении этих значений возрастает вероятность совершения мошеннической операции.
Положительные корреляции:
Признаки V7 и V20 имеют положительную корреляцию. То есть, с увеличением этих значений вероятность того, что операция окажется мошеннической, возрастает.

```python
plt.figure(figsize=(12, 9))
sns.boxplot( data = df1, x='Class', y='V2')
plt.title('V2 vs Class')
plt.show()
```
