```python
# Иванова Анастасия УБ22-08Б
# Вариант 7
# Задание 1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Генерация массива с равномерным распределением
uniform_array = np.random.uniform(7, 15, 6)

# Генерация трех массивов с нормальным распределением
norm_array1 = np.random.normal(8.0, 1.6, 6)
norm_array2 = np.random.normal(11.0, 1.2, 6)
norm_array3 = np.random.normal(14.0, 0.8, 6)

# Генерация двух массивов с экспоненциальным распределением
exp_array1 = np.random.exponential(1/1.90, 6)
exp_array2 = np.random.exponential(1/(1.90/2), 6)

print("Равномерное распределение: ", uniform_array)
print("Нормальное распределение 1: ", norm_array1)
print("Нормальное распределение 2: ", norm_array2)
print("Нормальное распределение 3: ", norm_array3)
print("Экспоненциальное распределение 1: ", exp_array1)
print("Экспоненциальное распределение 2: ", exp_array2)
```
