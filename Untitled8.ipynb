Информация о данных:
- Объекты (строки) - пассажиры Титаника
- Перечень признаков (столбцов):
1. Pclass - класс пассажира;
2. Name - полное имя пассажира;
3. Sex - пол пассажира;
4. Age - возраст пассажира;
5. SibSp - количество родственников супруг+братья\сестры на борту;
6. Parch - количество родственников родители+дети на борту;
7. Ticket - номер билета;
8. Fare - стоимость билета;
9. Cabin - номер каюты;
10. Embarked - порт отправления;
11. Survived - выжил пассажир или нет (целевая переменная).
