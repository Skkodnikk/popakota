```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('/content/Стоимость жилья в Красноярске (2).csv')
df
```

```python
df= df.drop('id', axis=1)
df
```

```python
df.info()
```

```python
df.isnull().sum().max()
```

Как видно из функции, в таблице нет пустых значений.

```python
X = df.drop('Цена', axis=1)
Y = df['Цена']
```

```python
X.info()
Y.info()
```

```python
X_train = X[0::1]
Y_train = Y[0::1]
X_test = X[0::1]
Y_test = Y[0::1]
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
w = model.coef_
b = model.intercept_
```

```python
w,b
```

```python
Y_train_prediction = model.predict(X_train)
Y_test_prediction = model.predict(X_test)
```

```python
print('Train MAE: ', mean_absolute_error(Y_train, Y_train_prediction))
print('Test MAE: ', mean_absolute_error(Y_test, Y_test_prediction))
print('Train MSE: ', mean_squared_error(Y_train, Y_train_prediction))
print('Test MSE: ', mean_squared_error(Y_test, Y_test_prediction))
```

Средняя абсолютная погрешность, является примерно 220 тысяч рублей, что является достаточно критичной погрешностью по сравнению со средним значением стоимости квартиры, которая составляет более 2 млн.руб.

```python
X_values = np. linspace(0, Y_test.shape[0], Y_test. shape[0])
plt.figure(figsize=(15, 6))
plt. scatter (X_values, Y_test_prediction, color='aqua', label='Предсказанные данные', edgecolor='k')
plt. scatter(X_values, Y_test, color='pink', label='Тестовые данные', edgecolor='k')
plt.title('Предсказания стоимости квартиры по методу линейной регрессии')
plt.ylabel('Тестовая и предсказанная стоимость, млн руб')
plt.xlabel('Квартиры')
```

По графику можем заметить, что предсказанния модели следуют за тестовыми данными, но со значительными отклонениями. Это подтверждается высокими значениями метрик MAE и MSE, которые были получены ранее.

```python
row_index = 669
row = X_test.iloc[row_index]

# Получаем стоимость квартиры с помощью модели
predicted_cost = model.predict(row.values.reshape(1, -1))

# Получаем изначальное значение стоимости квартиры
actual_cost = Y_test.iloc[row_index]

fig, ax = plt.subplots(figsize=(8, 6))

x = ['Предсказанная', 'Фактическая']
y = [predicted_cost[0], actual_cost]
ax.bar(x, y, color=['pink', 'silver'], edgecolor='k')

ax.set_title(f'Сравнение предсказанной и фактической стоимости квартиры #{row_index}')
ax.set_xlabel('Стоимость, млн руб')
ax.set_ylabel('Стоимость, млн руб')

plt.show()
```

на графике видно, что ошибка в строке 669 не является критичной.
