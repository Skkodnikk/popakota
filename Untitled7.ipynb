Функция describe() выводит статистическую информацию по числовым столбцам, такую как среднее, стандартное отклонение, минимум, максимум и квартили. Полезно для первичного анализа данных. Сравнение статистических данных с данными графика выбросов позволяет обнаружить выбросы и аномалии в данных.
describe(include='all') включает в себя также анализ нечисловых столбцов.
describe(include=['object']) выдает статистику только для столбцов с типом "object" (нечисловых).
